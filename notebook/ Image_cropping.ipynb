```python
import cv2
import os

video_names = ["DOWN_Light","FORWARD_Light","LEFT_Light","RIGHT_Light","UP_Light","DOWN_Dark","FORWARD_Dark","LEFT_Dark","RIGHT_Dark","UP_Dark"]

def save_all_frames(video_path, dir_path, basename, ext='jpg'):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        return

    os.makedirs(dir_path, exist_ok=True)
    base_path = os.path.join(dir_path, basename)

    digit = len(str(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))))

    n = 0

    # interval = 10
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if ret:
          # if frame_count % interval == 0:
            cv2.imwrite('{}_{}.{}'.format(base_path, str(n).zfill(digit), ext), frame)
            n += 1
        else:
            return
        frame_count += 1

for video_name in video_names:
  save_all_frames(f"gdrive/My Drive/{video_name}.mp4",f'gdrive/My Drive/data/temp/result/{video_name}_2024-09-17',video_name)
```
