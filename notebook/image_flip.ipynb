```python
import glob
import re
folder_names = glob.glob("./gdrive/My Drive/data/temp/result/*")
folder_names_splited = []
pattern = ".+2024-09-17$"
# pattern = ".+nothing$"
#pattern = ".+guu$|.+paa$"
for folder_name in folder_names:
  result = re.match(pattern,folder_name)
  if result:
    print(folder_name)
    folder_names_splited.append(folder_name.split("/")[-1]) #パスからフォルダ名のみを取り出している
folders = {} #フォルダごとに持っているファイルのパスを保存
for folder_name in folder_names_splited:
  file_paths = glob.glob(f"./gdrive/My Drive/data/temp/result/{folder_name}/*")
  folders[folder_name]=file_paths
```

```python
import cv2
import os
# def Flip_horizontal(input_path,save_path):
#   img = cv2.imread(input_path)
#   img_flip_horizontal = cv2.flip(img,1)
#   cv2.imwrite(save_path,img_flip_horizontal)

def rotate_image(input_path,save_path):
  img = cv2.imread(input_path)
  rotated_img = cv2.flip(img,1)
  cv2.imwrite(save_path,rotated_img)


def main():
  for folder_name in folder_names_splited:
    os.makedirs(f"./gdrive/My Drive/data/temp/result/{folder_name}_fliped",exist_ok=True)
    for input_path in folders[folder_name]:
          print(input_path.split("/")[-1].split(".")[0]+".jpg")
          output_path = f"./gdrive/My Drive/data/temp/result/{folder_name}_fliped/"+input_path.split("/")[-1].split(".")[0]+".jpg"
          # Flip_horizontal(input_path,output_path)
          rotate_image(input_path,output_path)

if __name__ == "__main__":
  main()
```
