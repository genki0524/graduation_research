```python
import os
import glob
import random
import shutil
import re

def make_dataset(dir,train_output_dir,test_output_dir):
  file_num = sum(os.path.isfile(os.path.join(dir, name)) for name in os.listdir(dir))
  files = glob.glob(os.path.join(dir,"*"))
  random.shuffle(files)
  train_files = files[:int(file_num*0.8)]
  test_files = files[int(file_num*0.8):]
  for train_file in train_files:
    shutil.copy(train_file,train_output_dir)
  for test_file in test_files:
    shutil.copy(test_file,test_output_dir)
```

```python
folders = glob.glob("./gdrive/My Drive/data/temp/result/*")
pattern = ".+224x224$"
folders = [folder for folder in folders if re.match(pattern,folder)]
print(folders)
for folder in folders:
  folder_name = folder.split("/")[-1]
  train_output = os.path.join("./gdrive/My Drive/data/datasets/train/",folder_name)
  test_output = os.path.join("./gdrive/My Drive/data/datasets/test/",folder_name)
  os.makedirs(train_output)
  os.makedirs(test_output)
  make_dataset(folder,train_output,test_output)
```
